# Longitudinal Analysis — 2 Paddler Configuration

Adjust paddler positions and weights below, then click **Run Analysis**.

```python
import nbformat, os

# Read the original notebook and extract code cell sources
_nb_path = os.path.join(os.path.abspath(''), 'Longitudinal Analysis_2 Paddler.ipynb')
with open(_nb_path, 'r') as _f:
    _nb = nbformat.read(_f, as_version=4)

cell_sources = []
for _cell in _nb.cells:
    if _cell.cell_type == 'code':
        cell_sources.append(_cell.source)

# Execute cell 0: loads imports, constants, CSV data, and default inputs
get_ipython().run_cell(cell_sources[0])
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

weight_options = [(f'{w} kg', w) for w in pad_weights]

pad1_pos_slider = widgets.IntSlider(
    value=1750, min=0, max=int(max(x_out)), step=50,
    description='Paddler 1 Pos (mm):',
    style={'description_width': '160px'},
    layout=widgets.Layout(width='500px'),
    continuous_update=False
)
pad2_pos_slider = widgets.IntSlider(
    value=4150, min=0, max=int(max(x_out)), step=50,
    description='Paddler 2 Pos (mm):',
    style={'description_width': '160px'},
    layout=widgets.Layout(width='500px'),
    continuous_update=False
)
pad1_weight_dd = widgets.Dropdown(
    options=weight_options, value=pad_weights[0],
    description='Paddler 1 Weight:',
    style={'description_width': '160px'},
    layout=widgets.Layout(width='300px')
)
pad2_weight_dd = widgets.Dropdown(
    options=weight_options, value=pad_weights[1],
    description='Paddler 2 Weight:',
    style={'description_width': '160px'},
    layout=widgets.Layout(width='300px')
)
run_button = widgets.Button(
    description='Run Analysis',
    button_style='primary',
    layout=widgets.Layout(width='200px', height='40px')
)

controls = widgets.VBox([
    widgets.HBox([pad1_pos_slider, pad1_weight_dd]),
    widgets.HBox([pad2_pos_slider, pad2_weight_dd]),
    run_button
])
output_area = widgets.Output()
display(controls, output_area)
```

```python
def on_run_clicked(_):
    with output_area:
        clear_output(wait=True)
        print('Running analysis \u2026')

    # Override input variables from widgets
    ns = get_ipython().user_ns
    ns['pad1_pos'] = pad1_pos_slider.value
    ns['pad2_pos'] = pad2_pos_slider.value
    ns['pad_first'] = pad1_weight_dd.value
    ns['pad_second'] = pad2_weight_dd.value
    ns['pad1'] = ns['pad_first'] * ns['LLF']
    ns['pad2'] = ns['pad_second'] * ns['LLF']
    ns['total_mass'] = ns['canoe_mass'] + ns['pad1'] + ns['pad2']

    # Re-execute cells 1+ from the original notebook inside the output widget
    with output_area:
        clear_output(wait=True)
        ip = get_ipython()
        for src in cell_sources[1:]:
            ip.run_cell(src)

run_button.on_click(on_run_clicked)
```
